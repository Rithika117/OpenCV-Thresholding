Write a Python program using OpenCV to:

1. Read an image (road_lanes.png) and convert it to **grayscale**.
2. Apply **Global Thresholding** using `cv2.threshold()` with a fixed threshold value.
3. Display the **original grayscale image** and the **global thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
# Import
import cv2
import matplotlib.pyplot as plt


```

```python
# Read the image using OpenCV

image = cv2.imread('road_lanes.png', 0)
```

```python
ret, threshold_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
# Display the image using Matplotlib

# Display images using Matplotlib
plt.figure(figsize=(10, 5))

# Original Grayscale Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Thresholded Image
plt.subplot(1, 2, 2)
plt.imshow(threshold_image, cmap='gray')
plt.title('Global Thresholded Image')
plt.axis('off')

plt.show()
```



Write a Python program using OpenCV to:

1. Read an image (Piano_Sheet_Music.png) and convert it to **grayscale**.
2. Apply **Adaptive Thresholding** (Mean or Gaussian) using `cv2.adaptiveThreshold()`.
3. Display the **original grayscale image** and the **adaptive thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
img = cv2.imread("Piano_Sheet_Music.png", 0)
```

```python
adaptive_thresholded = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.subplot(122)
plt.imshow(adaptive_thresholded, cmap='gray')
plt.title("Adaptive Thresholded Image")
plt.axis('off')

plt.show()
```


Write a Python program using OpenCV to:

1. Read an image (road_lanes.png) and convert it to **grayscale**.
2. Apply **Otsu’s Thresholding** using `cv2.threshold()` with the `cv2.THRESH_OTSU` flag.
3. Display the **original grayscale image** and the **Otsu-thresholded image** using Matplotlib in a **1 × 2 layout**, each with an appropriate **title**.


```python
image = cv2.imread("road_lanes.png", 0)
```

```python
otsu_thresholded = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("road_lanes.png", 0)

ret, otsu_thresholded = cv2.threshold(
    image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(otsu_thresholded, cmap="gray")
plt.title("Otsu Thresholded Image")
plt.axis("off")

plt.show()
```
